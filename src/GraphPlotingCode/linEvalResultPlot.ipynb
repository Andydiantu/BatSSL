```python
import numpy as np

supervised_data = supervised  # Replace [...] with a list of lists or 2D array of data for supervised
SimClR_data = simCLR  # Replace [...] with a list of lists or 2D array of data for SimClR
autoEncoder_data = autoencoder  # Replace [...] with a list of lists or 2D array of data for autoEncoder
Random_data = ranmdom  # Replace [...] with a list of lists or 2D array of data for Random

supervised_data = np.array(supervised_data)
SimClR_data = np.array(SimClR_data)
autoEncoder_data = np.array(autoEncoder_data)
Random_data = np.array(Random_data)

print(supervised_data.shape)

supervised_mean = np.mean(supervised_data, axis=1)
SimClR_mean = np.mean(SimClR_data, axis=1)
autoEncoder_mean = np.mean(autoEncoder_data, axis=1)
Random_mean = np.mean(Random_data, axis=1)

supervised_std = np.std(supervised_data, axis=1)
SimClR_std = np.std(SimClR_data, axis=1)
autoEncoder_std = np.std(autoEncoder_data, axis=1)
Random_std = np.std(Random_data, axis=1)

# If you want to calculate the standard error of the mean (SEM), use:
supervised_sem = supervised_std / np.sqrt(supervised_data.shape[0])
SimClR_sem = SimClR_std / np.sqrt(SimClR_data.shape[0])
autoEncoder_sem = autoEncoder_std / np.sqrt(autoEncoder_data.shape[0])
Random_sem = Random_std / np.sqrt(Random_data.shape[0])
```

```python
plt.figure(figsize=(10, 7))
data = ['1%', '5%', '10%', '50%', '100%']
x_values = list(range(5))
plt.errorbar(x_values, supervised_mean, yerr=supervised_std, color='green', linestyle='-', marker='*', capsize=5, label='supervised')
plt.errorbar(x_values, SimClR_mean, yerr=SimClR_std, color='red', linestyle='--', marker='X', capsize=5, label='SimCLR')
plt.errorbar(x_values, autoEncoder_mean, yerr=autoEncoder_std, color='blue', linestyle='dashdot', marker='o', capsize=5, label='Autoencoder')
plt.errorbar(x_values, Random_mean, yerr=Random_std, color='black', linestyle='dotted', marker='D', capsize=5, label='Random')

plt.xticks(x_values, data, size=15)
plt.yticks(size=15)
plt.xlabel('Amount of Labbeled Data Used', size=15)
plt.ylabel('Macro Accuracy', size=15)
plt.legend(fontsize=15)
plt.savefig('logisticRegression_with_errorbars.pdf', dpi=300)
```
