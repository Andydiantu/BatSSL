```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.signal import spectrogram

# Read the audio file
sample_rate, audio_data = wavfile.read('example_audio/bat_data_martyn_2018/BARBAR_20180723_215600-BARBAR-LR_0_1.0.wav')

# Calculate the time axis for the waveform plot
time = np.arange(0, len(audio_data)) / sample_rate

# Calculate the spectrogram
frequencies, times, spectrogram_data = spectrogram(audio_data, fs=sample_rate)

# Create a 2-row, 1-column grid for the plots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 6), sharex=True)

# Plot the waveform on the top
ax1.plot(time, audio_data)
ax1.set_ylabel('Amplitude')
ax1.set_title('Waveform')

# Plot the spectrogram on the bottom
img = ax2.pcolormesh(times, frequencies, np.log(spectrogram_data), shading='auto', cmap='inferno')
ax2.set_ylabel('Frequency (Hz)')
ax2.set_xlabel('Time (s)')
ax2.set_title('Spectrogram')

# # Add a colorbar for the spectrogram
# cbar = fig.colorbar(img, ax=ax2)
# cbar.set_label('Power (log)')
plt.xlim(0, 1.0) 
print(left)
# Display the plots
plt.savefig('spectrogramExample.pdf', dpi = 20)
plt.show()
```
